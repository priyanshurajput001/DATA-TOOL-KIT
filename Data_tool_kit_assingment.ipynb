1. Create 2D NumPy array & sum of each row

```python
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sum = np.sum(arr, axis=1)
print("Row-wise sum:", row_sum)
```

```python
# 2. Pandas: Mean of a specific column
import pandas as pd

df = pd.DataFrame({'Marks': [70, 80, 90, 60]})
mean_value = df['Marks'].mean()
print("Mean of Marks column:", mean_value)

```

```python
# 3. Scatter plot using Matplotlib
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# 4. Correlation matrix with Seaborn (Heatmap)


import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'Math': [80, 85, 78],
    'Science': [90, 88, 92],
    'English': [75, 70, 80]
})

correlation = df.corr()
sns.heatmap(correlation, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 5. Bar plot using Plotly

import plotly.express as px
import pandas as pd

df = pd.DataFrame({'Subjects': ['Math', 'Science', 'English'], 'Marks': [80, 90, 70]})
fig = px.bar(df, x='Subjects', y='Marks', title="Bar Plot")
fig.show()
```

```python
# 6. Create DataFrame & add new column

import pandas as pd

df = pd.DataFrame({'Marks': [80, 90, 85]})
df['Grade'] = ['A', 'A+', 'B']
print(df)
```

```python
# 7. Element-wise multiplication of two NumPy arrays

import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a * b
print("Element-wise multiplication:", result)
```

```python
# 8. Line plot with multiple lines using Matplotlib

import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 30, 40]
y2 = [5, 15, 25, 35]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.title("Multiple Lines")
plt.show()
```

```python
# 9. Filter Pandas DataFrame rows based on column value

import pandas as pd

df = pd.DataFrame({'Marks': [80, 50, 90, 60]})
filtered_df = df[df['Marks'] > 60]
print(filtered_df)
```

```python
# 10. Histogram using Seaborn

import seaborn as sns
import matplotlib.pyplot as plt

data = [10, 20, 20, 30, 40, 40, 40, 50]
sns.histplot(data)
plt.title("Histogram")
plt.show()
```

```python
# 11. Matrix multiplication using NumPy

import numpy as np

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
result = np.dot(a, b)
print("Matrix multiplication:\n", result)
```

```python
# 13. 3D Scatter plot using Plotly

import plotly.express as px
import pandas as pd

df = pd.DataFrame({
    'x': [1, 2, 3],
    'y': [4, 5, 6],
    'z': [7, 8, 9]
})

fig = px.scatter_3d(df, x='x', y='y', z='z', title="3D Scatter Plot")
fig.show()
```

1.	What is NumPy, and why is it widely used in Python?
NumPy is a library for numerical computing. It's fast and great for working with arrays and matrices.
2.	How does broadcasting work in NumPy?
Broadcasting lets NumPy perform operations on arrays of different shapes by automatically adjusting their shapes.
3.	What is a Pandas DataFrame?
A DataFrame is a 2D table with rows and columns, like an Excel sheet, used for storing and analyzing data.
4.	Explain the use of the groupby() method in Pandas.
groupby() is used to split data into groups, apply functions (like sum or mean), and combine results.
5.	Why is Seaborn preferred for statistical visualizations?
Seaborn makes beautiful, informative plots easily and is built on top of Matplotlib with more features.
6.	What are the differences between NumPy arrays and Python lists?
NumPy arrays are faster, support vector operations, and use less memory than Python lists.
7.	What is a heatmap, and when should it be used?
A heatmap shows data using colors. Use it to visualize patterns or correlations in a matrix.
8.	What does the term 'vectorized operation' mean in NumPy?
It means applying operations to entire arrays without writing loops, which makes code faster.
9.	How does Matplotlib differ from Plotly?
Matplotlib is static and script-based; Plotly is interactive and web-based.
10.	What is the significance of hierarchical indexing in Pandas?
It allows multiple levels of indexing for rows and columns, helping to work with complex data.
11.	What is the role of Seaborn's pairplot() function?
It shows pairwise relationships in data with scatter plots and histograms.
12.	What is the purpose of the describe() function in Pandas?
describe() gives quick stats like mean, count, min, and max for numeric columns.
13.	Why is handling missing data important in Pandas?
Missing data can affect results, so we handle it to keep analysis accurate.
14.	What are the benefits of using Plotly for data visualization?
Plotly makes interactive, attractive charts with zoom, hover, and clickable features.
15.	How does NumPy handle multidimensional arrays?
NumPy uses n-dimensional arrays (ndarray) to work with grids of data efficiently.
16.	What is the role of Bokeh in data visualization?
Bokeh creates interactive plots for web use, good for dashboards and live data.
17.	Explain the difference between apply() and map() in Pandas.
map() is for element-wise changes in a Series. apply() is more flexible and works on Series or
DataFrames.
18.	What are some advanced features of NumPy?
Advanced indexing, broadcasting, linear algebra, FFT, random sampling.
19.	How does Pandas simplify time series analysis?
Pandas handles date/time data well with easy resampling, rolling windows, and date indexing.
20.	What is the role of a pivot table in Pandas?
Pivot tables summarize and rearrange data for better analysis.
21.	Why is NumPy's array slicing faster than Python's list slicing?
NumPy arrays are stored in continuous memory blocks and processed in C, making slicing faster.
22.	What are some common use cases for Seaborn?
Distribution plots, regression plots, heatmaps, and visualizing categorical data.
